## Import Data Set:

#### Scatter plot with regression line(by default)

```python
# we import seaborn,visualization library
import seaborn as sns;
sns.set(color_codes=True)
import matplotlib.pyplot as plt;
import pandas as pd;

# Load CSV and columns
chirp_data = pd.read_csv("অংকে মেশিন লার্নিং ২, ঝিঁঝিঁ পোকার তাপমাত্রা DataSet/chirps.csv")

g = sns.lmplot(x="Chirps_15s", y="Temp_C", data=chirp_data)
plt.title('Cricket Chirp and Temperature')
plt.show()
```

#### Scatter plot without regression line.

```python
# we import seaborn,visualization library
import seaborn as sns;
sns.set(color_codes=True)
import matplotlib.pyplot as plt;
import pandas as pd;

# Load CSV and columns
chirp_data = pd.read_csv("অংকে মেশিন লার্নিং ২, ঝিঁঝিঁ পোকার তাপমাত্রা DataSet/chirps.csv")

g = sns.lmplot(x="Chirps_15s", y="Temp_C",fit_reg = False, data=chirp_data)
plt.title('Cricket Chirp and Temperature')
plt.show()
```
